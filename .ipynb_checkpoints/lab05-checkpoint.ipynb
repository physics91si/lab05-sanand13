# Part 1

```python
def integrate(y, dx):
    """
    Takes an array of y-values and a step size dx of the corresponding x values 
    and numerically integrates the function, returning the final value
    """
    counter = 0
    for yi in y:
        counter += dx * yi
    return counter
```

```python
def plot_fn(xarr, yarr, title=''):
    """
    Basic plotting function
    """
    plt.plot(xarr, yarr)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
# TODO write code here to setup arrays x and y = sin(x) and then plot them by calling plot_fn
xarr = np.arange(0, 2*math.pi, 0.01)
xlist = xarr.tolist()
ylist = list(map(np.sin, xlist))
yarr = np.asarray(ylist)

plot_fn(xarr, yarr, title = 'Sin(x)')
```

```python
# TODO then integrate y 

print (integrate(yarr, 0.01))
print (integrate(yarr, 0.001))
print (integrate(yarr, 0.0001))
```

The value of the integral gets closer to 0 as we increase the step size.

```python
# TODO find local maxima and minima of sinx from 0 to 6*pi

xarr = np.arange(0, 6*math.pi, 0.01)
xlist = xarr.tolist()
ylist = list(map(np.sin, xlist))
yarr = np.asarray(ylist)

plt.plot(xarr, yarr, label = "sin")

ydiff = np.diff(yarr)
ydiffsign = np.sign(ydiff)

maxima = []
minima = []
for n in range (0, len(ydiffsign)-1):
    if ydiffsign[n] == -1 and ydiffsign[n+1] == 1:
        minima.append((xarr[n]+xarr[n+1])/2)
    if ydiffsign[n] == 1 and ydiffsign[n+1] == -1:
        maxima.append((xarr[n]+xarr[n+1])/2)

print ('Local Minima found at',minima)
print ('Local Maxima found at',maxima)
```

The locations of the local extrema are close to what we expect.

```python
def wavepacket(x, k, sigma):
    """
    This function creates a wavepacket on the interval defined by x with
    wavevector k and standard deviation sigma.
    """
    return np.sin(k*x) *  np.exp(-(x**2)/(2*sigma**2))
```

```python
def noisy_packet(x_values, k, sigma, noise_amplitude):
    """
    This function returns a noisy Gaussian wavepacket with wave
    vector k, standard deviation sigma and Gaussian noise of standard
    deviation noise_amplitude.
    """
    clean_y = wavepacket(x_values,k,sigma)
    noisy_y = clean_y + noise_amplitude*np.random.randn(len(x_values))
    return noisy_y
```

```python
xarr = np.arange(0, 2*math.pi, 0.01)

f, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharex = True, sharey = True)
ax1.plot(xarr, noisy_packet(xarr, 1, 1, 0.1))
ax2.plot(xarr, noisy_packet(xarr, 2, 1, 0.1))
ax3.plot(xarr, noisy_packet(xarr, 1, 10, 0.1))
ax4.plot(xarr, noisy_packet(xarr, 1, 1, 0.2))
plt.show()
```

```python
yarr = noisy_packet(xarr, 5, 1, 0.2)
plt.plot(xarr, yarr)
plt.show()
```

The plot looks like what we expect, as we seem to have 5 oscillations (wave number k is 5). And the noise amplitude seems to be around 0.2.

```python
def clean_data(x_values,y_values):
    """
    This function should take a set of y_values, perform the Fourier
    transform on it, filter out the high frequency noise, transform the
    signal back into real space, and return it.
    """    
    # TODO edit this function
    y_fft = np.fft.rfft(y_values)
    
    low_pass_filter = np.ones(y_fft.shape) # build low pass filter for Fourier function
    num_freq = y_fft.size
    low_pass_filter[int(num_freq/25):num_freq] = 0
    
    y_fft_clean = []
    for i,e in enumerate(low_pass_filter):
        if e == 1:
            y_fft_clean.append(y_fft[i])
        elif e == 0:
            y_fft_clean.append(0)
    
    y_clean_values = np.fft.irfft(y_fft_clean)
    
    return y_clean_values
```

```python
# TODO call noisy_packet() to get a Gaussian wave packet, 
xarr = np.arange(0, 2*math.pi, 0.01)
yarr = noisy_packet(xarr, 5, 1, 0.2)

# call clean_data() to apply a low pass filter to the data, and
yclean = clean_data(xarr, yarr)

xclean = []
for i in range(0,len(yclean)):
    xclean.append(xarr[i])

# finally plot the result using plot_fn
plt.plot(xarr, yarr)
plt.plot(xclean, yclean)
plt.show()
```
